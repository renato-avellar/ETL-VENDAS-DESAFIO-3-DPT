```python
import duckdb
import os

# 1. Verificação de Arquivos
arquivos = ['modelagem_vendas.sql', 'electronics_sales_silver.csv']
for f in arquivos:
    if not os.path.exists(f):
        print(f"❌ Faltando: {f}")
```

```python
# 2. Criar Banco de Dados
DB_NAME = 'vendas_analitico.db'
if os.path.exists(DB_NAME): os.remove(DB_NAME)

con = duckdb.connect(DB_NAME)
with open('modelagem_vendas.sql', 'r') as f:
    con.execute(f.read())

print("✅ Banco de Dados Criado!")
```

```python
# 3. Validar Tabelas
con.execute("SHOW TABLES").df()
```
