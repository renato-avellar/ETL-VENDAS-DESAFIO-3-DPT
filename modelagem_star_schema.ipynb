```python
import duckdb
```

```python
con = duckdb.connect('vendas.db')
```

```python
with open('modelagem_vendas.sql', 'r') as f:
  sql_script = f.read()

con.execute(sql_script)
```

```python
df_teste = con.execute("""
        SELECT
            p.brand,
            SUM(f.net_revenue) as total_vendas
        FROM Fato_Vendas f
        JOIN Dim_Produto p ON f.product_id = p.product_id
        GROUP BY p.brand
        ORDER BY total_vendas DESC
        LIMIT 5
    """).df()

display(df_teste)
```
