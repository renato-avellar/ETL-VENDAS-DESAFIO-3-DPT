```python
# 2. Configuração do Banco de Dados
DB_NAME = 'vendas_analitico.db'

con = duckdb.connect(DB_NAME)
```

```python
with open('modelagem_vendas.sql', 'r', encoding='utf-8') as f:
  sql_script = f.read()

con.execute(sql_script)
```

```python
display(con.execute("SHOW TABLES").df())
```

```python
query = """
SELECT
    p.category,
    p.brand,
    SUM(f.net_revenue) as receita_liquida,
    SUM(f.gross_profit) as lucro_bruto
FROM Fato_Vendas f
JOIN Dim_Produto p ON f.product_id = p.product_id
GROUP BY ALL
ORDER BY lucro_bruto DESC
LIMIT 10
"""

display(con.execute(query).df())
```
